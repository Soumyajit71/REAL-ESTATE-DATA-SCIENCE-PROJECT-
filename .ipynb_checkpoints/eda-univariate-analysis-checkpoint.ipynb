# Price

```python
df['price'].isnull().sum()
```

```python
df['price'].describe()
```

```python
sns.histplot(df['price'],kde=True,bins=50)
```

```python
sns.boxplot(x=df['price'],color='lightgreen')
```

### Descriptive Statistics 
    a. Count: There are 3,660 non missing values.
    b. Mean Price: The average price is approximately 2.53 crores.
    c. Median Price: The median price(or 50 th percentile) is 1.52 crores.
    d. Standard deviation : The prices have a standard deviation of 2.98, including variability in the prices
    e. Range: Price range from a minimum of 0.07 crores to a maximum of 31.5 crores.
    f. IQR: The interquartile range (diff b/w 75th and 25th percentile) is from 0.95 crores to 2.75 crores
### Visualizations:
    a. Distribution: The histogram indicates that most properties are priced in the lower range (below 5 crores), with a few 
    properties going beyond 10 crores.
    b. Box Plot: Box plot shows the spread of the data and potential outliers.properties price above 10 crores might be 
    considered outliers as they lie beyond the upper whisker of the box plot.
### Missing Values:
    There are 17 missing values in the price column

```python
# skewness and kurtosis

skewness=df['price'].skew()
kurtosis=df['price'].kurt()

print(skewness,kurtosis)
```

### Skewness: 
    The price distribution has a skewness of approximately 3.28, indicating a positive skew. This means that distribution 
    tail is skewed to the right,which aligns with our observation from the histogram where most properties have prices on  
    the lower end with few high-priced properties.
### Kurtosis:
    The Kurtosis value is approximately 14.93. A kurtosis value greater than 3 indicates that with havier tails and more
     outliers compared to normal distribution

```python
# Quantile Analysis
quantiles=df['price'].quantile([0.01, 0.05, 0.95, 0.99])

quantiles
```

### Quantile Analysis:
     a. 1% Quantile: Only 1% properties are priced below 0.25 crores.
     b. 5% Quantile: Only 5% properties are priced below 0.37 crores.
     c. 95% Quantile: 95% properties are priced below 8.5 crores.
     d. 99% Quantile: 99% properties are priced below 15.26 crores, indicating that there are very few properties whose 
     prices are above 15.26 crores.

```python
# Identifying potential outliers using IQR method
Q1=df['price'].describe()['25%']
Q3=df['price'].describe()['75%']

IQR=Q3-Q1
```

```python
lower_bound= Q1 - 1.5 * IQR
upper_bound= Q3 + 1.5 * IQR

print(lower_bound,upper_bound)
```

```python
outliers=df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]
```

```python
outliers.shape
```

```python
outliers['price'].describe()
```

### Outliers Analysis(using IQR method)
    a. based on the IQR method, there are 425 properties are considered as outliers.
    b. These outliers have an average price of approximately 9.24 crores.
    c. The range for these outliers is from 5.46 crores to 31.5 crores.

```python
# price binning
bins=[0,1,2,3,5,10,20,50]
labels=['0-1','1-2','2-3','3-5','5-10','10-20','20-50']
pd.cut(df['price'],bins=bins,labels=labels,right=False).value_counts().sort_index().plot(kind='bar')
```

a. The majority of properties are priced in the "1-2 crores" and "2-3 crores" ranges.

b. There is a significant drop in the number of properties priced above "5 crores".

```python
# ecdf plot
ecdf=df['price'].value_counts().sort_index().cumsum()/len(df['price'])
plt.plot(ecdf.index,ecdf,marker='.',linestyle='none')
```

```python
plt.figure(figsize=(15,6))

# Distribution plot without log transformation
plt.subplot(1,2,1)
sns.histplot(df['price'],kde=True,bins=50,color='skyblue')
plt.title('Distribution of Prices (Orginal)')
plt.xlabel('Prices (in Crores)')
plt.ylabel('Frequency ')

# Distribution plot with log transformation
plt.subplot(1,2,2)
sns.histplot(np.log1p(df['price']),kde=True,bins=50,color='lightgreen')
plt.title('Distribution of Prices (Orginal)')
plt.xlabel('Prices (in Crores)')
plt.ylabel('Frequency ')
```

np.log1p(x): This function computes the natural logarithm of 1+x.It's designed to provide more accurate results for values of x 
that are very close to zero.

Using np.log1p helps in transforming the price column while enssuring that any value (including zero, if presents) is handled appropriately.
When we need to reverse the transformation, we can use np.expm1 which computes e^x-1

```python
skewness=np.log1p(df['price']).skew()
kurtosis=np.log1p(df['price']).kurt()
print(skewness,kurtosis)
```

```python
plt.figure(figsize=(15,6))

# Box Plot without log transformation 
plt.subplot(1,2,1)
sns.boxplot(df['price'],color='skyblue')
plt.title('Distribution of Prices (Original)')
plt.xlabel('Price (in Crores)')
plt.ylabel('Frequency')

# Box plot with log transformation

plt.subplot(1,2,2)
sns.boxplot(np.log1p(df['price']),color='lightgreen')
plt.title('Distribution of Prices (log transformed)')
plt.xlabel('Price (in Crores)')
plt.ylabel('Frequency')
```
